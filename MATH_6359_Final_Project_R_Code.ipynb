# Q3 - What features can determine diabetes rate?

````r
## Q3 - factors that affect health (diabetes rates)
#install.packages("GGally")
#install.packages("corrplot")
#install.packages("randomForest")
library(GGally)
library(corrplot)
library(randomForest)

fea = read.csv("project_dataset_v3.csv")

head(fea)

fea$METRO13 = as.factor(fea$METRO13)


summary(fea)

fea_clean <- na.omit(fea)
summary(fea_clean)
dim(fea_clean)

ggpairs(fea_clean[4:16])

#transform some variables by log

fea_transformed = fea_clean

fea_transformed$FFRPTH11 = log(fea_transformed$FFRPTH11)
fea_transformed$FSRPTH11 = log(fea_transformed$FSRPTH11)
fea_transformed$RECFACPTH11 = log(fea_transformed$RECFACPTH11)
fea_transformed$PC_DIRSALES12 = log(fea_transformed$PC_DIRSALES12)
fea_transformed$PCT_LACCESS_POP10 = sqrt(fea_transformed$PCT_LACCESS_POP10)
fea_transformed$GROCPTH11 = log(fea_transformed$GROCPTH11)

ggpairs(fea_transformed[4:16])

df_final = fea_transformed[4:16]

ggpairs(df_final)

head(df_final)
````

````r
## Multiple linear regression

df = fea_clean[4:16]

#mlr = lm(df_final$PCT_DIABETES_ADULTS13~., data=df_final)
mlr = lm(fea_clean[4:16]$PCT_DIABETES_ADULTS13~., data=fea_clean[4:16])
summary(mlr)

# stepwise regression
start.model = lm(fea_clean[4:16]$PCT_DIABETES_ADULTS13 ~ 1, data = fea_clean[4:16])
step.backward = step(mlr, direction = "backward")
step.forward = step(start.model, direction = "forward", scope =
                      formula(mlr))
step.both = step(mlr, direction = "both")
step.backward
step.forward
step.both


# after stepwise
mlr_2 = lm(df$PCT_DIABETES_ADULTS13~ FFRPTH11 + FSRPTH11 + 
             MEDHHINC15 + PCT_65OLDER10 + METRO13 + RECFACPTH11 + PC_DIRSALES12 + 
             PCT_LACCESS_POP10 + PC_SNAPBEN12, data=df)
summary(mlr_2)
````

````r
## Random Forest

df = fea_clean[4:16]

set.seed(1)

#fit the random forest model
rf <- randomForest(
  formula = df$PCT_DIABETES_ADULTS13 ~ .,
  data = df
)

#display fitted model
rf

#find number of trees that produce lowest test MSE
which.min(rf$mse)

#find RMSE of best model
sqrt(rf$mse[which.min(rf$mse)]) 

#plot the test MSE by number of trees
plot(rf)

#produce variable importance plot
varImpPlot(rf) 
````

# Q4 - Can we predict the amount of SNAP benefits per capita?

### Initial prep

````r
project_dataset_v2 <- read.csv("project_dataset_v2.csv")
project_dataset_v2 = na.omit(project_dataset_v2)
df_no_state_county = project_dataset_v2[,c(-1,-2,-3)]
df_state_county = project_dataset_v2[,-1]
````

### No State/County Models

#### Multiple Linear Regression Model

##### Initial Model

````r
model = lm(PC_SNAPBEN12~., data = df_no_state_county)
par(mfrow = c(2,2))
plot(model)
summary(model)
````

##### Determining best variables using Stepwise function

````r
step(model)
````

##### Multiple Linear Regression Model from Stepwise process

````r
model = lm(formula = PC_SNAPBEN12 ~ PCT_DIABETES_ADULTS13 + FFRPTH11 + 
    POVRATE15 + MEDHHINC15 + PCT_65OLDER10 + PCT_18YOUNGER10 + 
    METRO13 + RECFACPTH11 + PCT_LACCESS_POP10 + GROCPTH11, data = df_no_state_county)
par(mfrow = c(2,2))
plot(model)
summary(model)
````

#### Random Forest

##### Train/Test Split

````r
set.seed(123)
sample = sample.int(n = nrow(df_no_state_county), size= floor(0.8*nrow(df_no_state_county)))
train = df_no_state_county[sample,]
test = df_no_state_county[-sample,]
````

##### RF Model

````r
library(randomForest)
y.test = test[,13]
X.test = test[,-13]
n = nrow(df_no_state_county)
p = ncol(df_no_state_county)-1
set.seed(123)
rf.df_no_state_county = randomForest(PC_SNAPBEN12 ~ .,
                                     data = train,
                                     xtest = X.test,
                                     ytest = y.test,
                                     ntree = 1000,
                                     mtry= sqrt(p),
                                     importance = TRUE)
rf.df_no_state_county
````

##### Variable Importances Plots

````r
varImpPlot(rf.df_no_state_county)
````
